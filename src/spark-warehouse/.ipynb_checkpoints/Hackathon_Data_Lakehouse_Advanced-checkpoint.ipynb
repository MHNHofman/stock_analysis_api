### Create a Spark session

✏️ 1. Initialize your **credentials** and specify **Spark session** arguments:

````python
# Initialize your creds and specify spark session arguments
import yaml

with open('secrets.yaml', 'r') as file:
    secrets = yaml.safe_load(file)
    
ACCESS_KEY = secrets["MINIO_ACCESS_KEY"]
SECRET_KEY = secrets["MINIO_SECRET_KEY"]
APP_NAME = f"hackathon_app_{your_initials}"
CATALOG_NAME = f"hackathon_catalog_{your_initials}"
````

✏️ 2. Initialize a **Spark session**:

````python
# Import packages
import os
import socket
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Set configs for Spark Session
spark_executor_cores = "2"
maven_packages = [
    "org.apache.iceberg:iceberg-spark-runtime-3.3_2.12:1.4.2",
    "org.apache.hadoop:hadoop-aws:3.3.2"
]

# Additional connection logic
def get_pod_ip():
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        s.connect(("8.8.8.8", 80))
        ip = s.getsockname()[0]
        s.close()
        return ip
    except Exception as e:
        return f"Error occurred: {e}"

catalog_name = CATALOG_NAME
hive_url = "thrift://hive-metastore-nl-main-dev-dev:9083"

# Create session
spark = (SparkSession
        .builder
        .master(os.getenv("SPARK_MASTER_URL"))
        .config("spark.driver.host", get_pod_ip())
        .config("spark.jars.packages", ",".join(maven_packages))
        .config("spark.hadoop.fs.s3a.endpoint", "https://minio-api.dev.nl.logex.cloud")
        .config("spark.hadoop.fs.s3a.access.key", ACCESS_KEY)
        .config("spark.hadoop.fs.s3a.secret.key", SECRET_KEY)
        .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "true")
        .config("spark.hadoop.fs.s3a.path.style.access", "true")
        .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
         
        .config("spark.sql.defaultCatalog", catalog_name)
        .config(f"spark.sql.catalog.{catalog_name}", "org.apache.iceberg.spark.SparkCatalog")
        .config(f"spark.sql.catalog.{catalog_name}.type", "hive")
        .config(f"spark.sql.catalog.{catalog_name}.uri", hive_url)
        .config(f"spark.sql.catalog.{catalog_name}.warehouse", "s3a://test.warehouse.perform")
        .config("spark.sql.catalogImplementation", "hive")
        .config("spark.cores.max", 2)
        .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions")
        
        .appName(APP_NAME)
        .enableHiveSupport()
        .getOrCreate())
````

## 1. Create Spark dataframes

✏️ Complete the following questions, tasks:
+ Go through the code below. What Spark **data types** do you know?
+ Do you know other ways **Spark dataframes** can be **created**?
+ **Add a new row** with *patient* data to tables `patients_data` list and `patient_info_data` list

````python
from pyspark.sql.types import *
from pyspark.sql import SparkSession

# Create Schemas for tables
patients_schema = StructType([
    StructField("patient_id", StringType()),
    StructField("activity_id", StringType()),
    StructField("activity_start_date", StringType()),
    StructField("activity_end_date", StringType()),
    StructField("doctor_id", StringType())])

patients_info_schema = StructType([
    StructField("patient_id", StringType()),
    StructField("patient_age", IntegerType()),
    StructField("bad_habits", ArrayType(StringType()))])

doctors_schema = StructType([
    StructField("doctor_id", StringType()),
    StructField("doctor_speciality", StringType()),
    StructField("doctor_experience_years", IntegerType())])

activities_schema = StructType([
    StructField("activity_id", StringType()),
    StructField("activity_description", StringType()),
    StructField("activity_type", StringType())])
````

````python
# Fill in data
patients_data = [("12", "01", "2023-01-12 03:14:07", "2023-01-12 13:14:07" , "doc_1"),
                 ("13", "02", "2023-01-12 10:14:07", "2023-01-12 16:14:07", "doc_2"),
                 ("11", "03", "2023-01-12 01:14:07", "2023-01-13 13:14:07", "doc_1"),
                 ("13", "01", "2023-01-12 17:14:07", "2023-01-12 18:14:07", "doc_3")]
                  
patients_info_data = [("11", 34, ["smoking", "drinking"]),
                      ("12", 56, []),
                      ("13", 28, ["drugs"])]   
                  
doctors_data = [("doc_1", "surgeon", 5), 
                ("doc_2", "intern", 1), 
                ("doc_3", "cardiologist", 6)]
                  
activities_data = [("01", "general health condition check", "therapeutic"), 
                   ("02", "blood analysis", "therapeutic"),
                   ("03", "heart surgery", "cardio")]
````

````python
# Create dataframes with schemas specified
               
patients_df =  spark.createDataFrame(patients_data,schema=patients_schema)
patients_info_df =  spark.createDataFrame(patients_info_data,schema=patients_info_schema)
doctors_df =  spark.createDataFrame(doctors_data,schema=doctors_schema)
activities_df =  spark.createDataFrame(activities_data,schema=activities_schema)

patients_df.show()
patients_info_df.show()
doctors_df.show()
activities_df.show()
````

## 2. Conditional JOIN exercise

### 2. Task 

✏️ Enrich patients data with patients info. **Add additional column** `risk_indicator` based on the length of `bad_habbits` array":
- `if len(habits_array) < 1` --> **"low"**
- `if len(habits_array) > 2` --> **"high"**
- `if len(habits_array) [1,2]` --> **"medium"**

````python
from pyspark.sql.functions import *

"""
Notice: WHEN-OTHERWISE is equivalent to IF-ELSE Python statement
OTHERWISE can have any additional conditions, or just be used as simple ELSE clause
"""
risk_indicator_col = when(size(col("bad_habits")) > 2, lit("high"))\
                     .otherwise(
                        when(size(col("bad_habits")) < 1, lit("low"))
                        .otherwise(lit("medium"))
                            )

patients_enriched_df = patients_df\
    .join(patients_info_df,
          patients_df.patient_id == patients_info_df.patient_id,
          'inner')\
    .withColumn("risk_indicator", risk_indicator_col)

"""
Check everything worked as expected with park show() command
"""
patients_enriched_df.show()
patients_enriched_df.printSchema()
````

### 2. Answer

2.1. Example of fixed JOIN statement:

````python
# Example of fixed JOIN statement 
patients_enriched_df = patients_df\
    .join(patients_info_df.withColumnRenamed("patient_id", "id"),
          col("patient_id")==col("id"))\
    .withColumn("risk_indicator", risk_indicator_col)\
    .drop(col("id"))

patients_enriched_df.show()
patients_enriched_df.printSchema()
````

2.2 Example of SQL syntax:

````python
# Example of SQL syntax

# Register the tables
patients_df.createOrReplaceTempView("patients_df")
patients_info_df.createOrReplaceTempView("patients_info_df")

# Run JOIN using SQL approach
patients_enriched_df_sql = spark.sql(
    """SELECT pat.patient_id, activity_id, activity_start_date, activity_end_date, doctor_id, bad_habits, patient_age,
    CASE WHEN size(bad_habits) < 1 THEN 'low' WHEN size(bad_habits) > 2 THEN 'high' ELSE 'medium' END as risk_indicator
    FROM patients_df pat
    JOIN patients_info_df info
    ON pat.patient_id = info.patient_id
    """).show()
````

## 3. Window functions exercise

✏️ Use **window functions** to check the previous patient's activity.  
Create column named `previous_activity` which will contain patient's previus activity.

````python
from pyspark.sql.window import Window
from pyspark.sql.functions import lag 

"""
Notice: it's more readable to have a window variable separately 
"""
window = Window.partitionBy("patient_id").orderBy("activity_start_date")
   
patients_enriched_df = patients_enriched_df.withColumn("previous_activity_id", lag("activity_id",1).over(window))
patients_enriched_df.show()
````

## 4. UDF example, implementation guide
1. Create python dictionaries from patient_info table and activity_info tables
2. Create function that takes all arguments needed for activities_info column, excluding activity_type (since it will be stored in dict made from activities_df)
3. Aggregate patients_activities tables by patient_id


✏️ What should be done to create the following dataframe?

```
patients_activities_info

|-- patient_id: string (nullable = true) \
|-- patient_age: integer (nullable = true) \
|-- risk_indicator: string (nullable = true) \
|-- bad_habits: array (nullable = true) \
|    |-- element: string (containsNull = true) \
|-- activities_info: array (nullable = true) \
|    |-- element: activity_id (containsNull = true) \
|    |-- element: activity_start_date (containsNull = true) \
|    |-- element: activity_end_date (containsNull = true) \
|    |-- element: previous_activity_id (containsNull = true) \
|    |-- element: activity_type (containsNull = true)
```
 
What can be some pitfalls when using UDFs?
Can it be done using standard pyspark functions?

````python
from pyspark.sql.functions import *

# Let's create a copy to implement udf function

patients_enriched_df_udf = patients_enriched_df

activity_info_dict = {}

for column in activities_df.columns: 
    activity_info_dict[column] = activities_df.toPandas()[column].values.tolist() 


def construct_activities_col(activity_id, previous_activity_id,
                            activity_start_date, activity_end_date):
    activity_type_index = activity_info_dict["activity_id"].index(activity_id)
    activity_type = activity_info_dict['activity_type'][activity_type_index]
    return {"activity_id": activity_id,
            "activity_start_date": activity_start_date,
            "activity_end_date": activity_end_date, 
            "previous_activity_id": previous_activity_id,
            "activity_type": activity_type
           }

# Create a schema for struct type column

constructUDF = udf(construct_activities_col,
                   StructType([
                       StructField("activity_id", StringType()),
                       StructField("previous_activity_id", StringType()),
                       StructField("activity_start_date", StringType()),
                       StructField("activity_end_date", StringType()),
                       StructField("activity_type", StringType()),
                   ]))

patients_enriched_df_udf = patients_enriched_df_udf\
.select("patient_id", "patient_age", "risk_indicator", "bad_habits",
        "activity_id", "activity_start_date", "activity_end_date","previous_activity_id",
        constructUDF(col("activity_id"), col("previous_activity_id"),
                     col("activity_start_date"), col("activity_end_date")
                                        ).alias("activities_info") ) \
# Check you can access struct field

patients_enriched_df_udf.select(col("activities_info.activity_id")).show()

# Create final aggregation, get rid of ambigous columns

# Implemnt aggregation with collect_list() agg. function
patients_enriched_df_udf = patients_enriched_df_udf\
.select("patient_id", "patient_age", "risk_indicator", "bad_habits", "activities_info") \
.groupby("patient_id", "patient_age", "risk_indicator", "bad_habits")\
.agg(collect_list("activities_info").alias("activities_info"))

# Does the schema look as expected?
patients_enriched_df_udf.printSchema()
````

✏️Step by step guide for non-UDF solution:
1. Enrich our df with all data needed 
2. Construct `activities_info` column 
3. Aggregate df by `patient_id`, creating aggreagated `activities_info` column
4. Select only column needed for `patients_activities_info` --> check the schema you have 

````python
# Let's create a copy to implement udf function

patients_enriched_df_simple = patients_enriched_df


# Enrich df with all data needed

patients_enriched_df_simple = patients_enriched_df_simple\
.join(activities_df.withColumnRenamed("activity_id", "id"), col("activity_id") == col("id"))\
.drop("info_id") \
.drop("id")


# Construct activities_info column
patients_enriched_df_simple = patients_enriched_df_simple\
.withColumn("activities_info", 
            struct("activity_id", 
                   "activity_start_date", 
                   "activity_end_date",
                   "previous_activity_id",
                   "activity_type"))

patients_enriched_df_simple.show()

# Implemnt aggregation with collect_list() agg. function
patients_enriched_df_simple = patients_enriched_df_simple\
.select("patient_id", "patient_age", "risk_indicator", "bad_habits", "activities_info") \
.groupby("patient_id", "patient_age", "risk_indicator", "bad_habits")\
.agg(collect_list("activities_info").alias("activities_info"))

# Does the schema look as expected?
patients_enriched_df_simple.printSchema()
````

## 5. Bonus track! 

✏️ Check the following solution which extends the current UDF implementation to act on array column of Struct type

````python
from pyspark.sql.functions import *

# Let's create a copy to implement udf function

patients_enriched_df_bonus = patients_enriched_df

activity_info_dict = {}

for column in activities_df.columns: 
    activity_info_dict[column] = activities_df.toPandas()[column].values.tolist() 

patients_enriched_df_bonus = patients_enriched_df_bonus\
.withColumn("activities_info_tmp",
            struct("activity_id", "activity_start_date", "activity_end_date","previous_activity_id")) \
.select("patient_id", "patient_age", "risk_indicator", "bad_habits", "activities_info_tmp") \
.groupby("patient_id", "patient_age", "risk_indicator", "bad_habits")\
.agg(collect_list("activities_info_tmp").alias("activities_info_tmp"))

# Check the schema you have now
patients_enriched_df_bonus.printSchema()


def construct_activities_col(activity_info_tmp):
    activity_info_array = []
    for item in activity_info_tmp:
        activity_id = item["activity_id"]
        activity_start_date = item["activity_start_date"]
        activity_end_date = item["activity_end_date"]
        previous_activity_id = item["previous_activity_id"]
        activity_type_index = activity_info_dict["activity_id"].index(activity_id)
        activity_type = activity_info_dict['activity_type'][activity_type_index]
        activity_info_array.append({"activity_id": activity_id,
            "activity_start_date": activity_start_date,
            "activity_end_date": activity_end_date, 
            "previous_activity_id": previous_activity_id,
            "activity_type": activity_type
           })
    return activity_info_array

# Create a schema for struct type column
constructUDF = udf(construct_activities_col,
                   ArrayType(StructType([
                       StructField("activity_id", StringType()),
                       StructField("previous_activity_id", StringType()),
                       StructField("activity_start_date", StringType()),
                       StructField("activity_end_date", StringType()),
                       StructField("activity_type", StringType()),
                   ])))

patients_enriched_df_bonus = patients_enriched_df_bonus\
.select("patient_id", "patient_age", "risk_indicator", "bad_habits","activities_info_tmp",
        constructUDF(col("activities_info_tmp")).alias("activities_info")).drop("activities_info_tmp")


# Does the schema look as expected?
patients_enriched_df_bonus.printSchema()
````
