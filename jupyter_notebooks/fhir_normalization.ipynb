```python
parquetExtension = spark.read.parquet('df.parquet_fhir_extension_total.gzip')  
parquetExtension.createOrReplaceTempView("parquetExtension")
extension = spark.sql("SELECT * FROM parquetExtension")
extension.show()
```

```python
parquetIdentifier = spark.read.parquet('df.parquet_fhir_identifier_total.gzip')  
parquetIdentifier.createOrReplaceTempView("parquetIdentfier")
identifier = spark.sql("SELECT * FROM parquetIdentfier")
identifier.show()
```
